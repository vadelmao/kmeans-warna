```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv("/content/drive/MyDrive/machinelearningaldo/dataset/color.csv")
rgb_data = data[['R', 'G', 'B']]  # Ambil kolom RGB
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(rgb_data['R'], rgb_data['G'], rgb_data['B'], color='gray', s=50)
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
plt.title("Plot Warna Sebelum Clustering")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(rgb_data)
data['Cluster'] = kmeans.labels_  # Tambah kolom cluster ke DataFrame
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data['R'], data['G'], data['B'],
                     c=data['Cluster'], cmap='viridis', s=50)
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
plt.title("Plot Warna Setelah Clustering (K-Means)")
plt.colorbar(scatter)
plt.show()
```
